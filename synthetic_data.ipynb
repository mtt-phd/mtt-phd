```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng(seed=42)
```

```python
num_timesteps = 10
```

```python
# starting coords; maybe change to random later on??
x_start = 0
y_start = 0
z_start = 0
```

```python
# random x, y, z velocities
dx = rng.uniform(0, 3)
dy = rng.uniform(0, 3)
dz = rng.uniform(0, 3)
print(dx, dy, dz)
```

```python
x = np.arange(x_start, num_timesteps * dx, dx)
y = np.arange(y_start, num_timesteps * dy, dy)
z = np.arange(z_start, num_timesteps * dz, dz)
coords = np.stack((x, y, z))
coords
```

```python
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(coords[0, :], coords[1, :], coords[2,:])
ax.title.set_text("3D trajectory")
```

```python
fig, axes = plt.subplots(3, sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()):
    ax.plot(np.arange(num_timesteps), coords[i,:])
    ax.title.set_text(f"{chr(ord('x') + i)}-pos")
```
