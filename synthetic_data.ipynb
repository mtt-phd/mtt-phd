```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.default_rng(seed=42)
```

```python
num_timesteps = 10
```

```python
# starting coords; maybe change to random later on??
x_start = rng.uniform(-2, 2)
y_start = rng.uniform(-2, 2)
z_start = rng.uniform(-2, 2)
print(x_start, y_start, z_start)
```

```python
# random x, y, z velocities
dx = rng.uniform(0, 3)
dy = rng.uniform(0, 3)
dz = rng.uniform(0, 3)
print(dx, dy, dz)
```

```python
x = np.linspace(x_start, dx * num_timesteps, num_timesteps)
y = np.linspace(y_start, dy * num_timesteps, num_timesteps)
z = np.linspace(z_start, dz * num_timesteps, num_timesteps)
print(x)
print(y)
print(z)
coords = np.stack((x, y, z))
coords
```

```python
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(coords[0, :], coords[1, :], coords[2,:])
ax.title.set_text("3D trajectory")
```

```python
fig, axes = plt.subplots(3, sharex=True, sharey=True)
for i, ax in enumerate(axes.flatten()):
    ax.plot(np.arange(num_timesteps), coords[i,:])
    ax.title.set_text(f"{chr(ord('x') + i)}-pos")
```
